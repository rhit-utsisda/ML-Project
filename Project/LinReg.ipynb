### Imports

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import LabelEncoder
```

```python
df = pd.read_csv("Datasets/KenPom Barttorvik.csv")
df.head()
```

```python
df.columns
```

```python
X = df.drop(['TEAM', 'WIN%', 'CONF','CONF ID','QUAD NO', 'QUAD ID','TEAM NO','TEAM ID','ROUND','YEAR','GAMES','W','L'], axis=1)  
y = df['WIN%']
X.head()
```

```python
X.columns
```

```python
X['3PTR']
```

```python
X = (X-X.mean())/X.std() #standardize it babyyyy
X.head()
```

```python

# label_encoder = LabelEncoder()
# X['Team_encoded'] = label_encoder.fit_transform(df['TEAM'])

# split into into training and testing set
X_train, X_test, y_train, y_test = train_test_split(X, y, train_size=0.8, random_state=0)

# Fit 
model = LinearRegression()
model.fit(X_train, y_train)
R2train = model.score(X_train, y_train)
print("Train R^2 score:", R2train)
# Score
R2 = model.score(X_test, y_test)
print("Test R^2 score:", R2)

```

```python
coefs = pd.Series(model.coef_, X.columns)
coefs
```
